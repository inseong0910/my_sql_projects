```python
plt.rcParams['font.family'] = 'Malgun Gothic'

conn = pymysql.connect(host='localhost', user='root', password='tpakdur8qjs@@', db='fastcampus_sql', charset='utf8')
cursor = conn.cursor()
```

```python
sql = "select substring(소재지전체주소, 1, instr(소재지전체주소,' ')), count(*) "
sql += " from camping_info"
sql += " group by substring(소재지전체주소, 1, instr(소재지전체주소,' '))"

cursor.execute(sql)
data = cursor.fetchall()

conn.close()
```

```python
labels = []
ratio = []

for (lb, cnt) in data:
    labels.append(lb)
    ratio.append(cnt)
    
plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
```
